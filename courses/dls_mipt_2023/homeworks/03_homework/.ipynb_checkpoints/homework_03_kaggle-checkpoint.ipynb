#### 1.1.2 Обработка значений

Задание функции, заменяющей пропуски на переданное значение и преобразующей значения к переданному типу данных:

````python
def replace_missed_vals_and_convert(df: pd.DataFrame, 
                                    col_name: str, 
                                    missed_val: str or int,
                                    fill_val: int or float,
                                    convert_to: str = None) -> pd.DataFrame:
    
    print(
        'Заменено пропущенных значений относительно всего набора данных:',
        round(df[df[col_name] == missed_val].shape[0] / df.shape[0] * 100, 2), '%'
    )
    
    df_updated = df.copy()
    
    # remove missed values
    # df_updated = df_updated[df_updated[col_name] != missed_val].reset_index(drop=True)
    
    # replace missed values
    df_updated[col_name] = df_updated[col_name].replace(missed_val, np.nan)
    df_updated[col_name] = df_updated[col_name].fillna(fill_val)
    
    # convert type
    if convert_to is not None:
        df_updated[col_name] = df_updated[col_name].astype(convert_to)
    
    return df_updated
````

Задание функции, приводящей строковые значения к нижнему регистру:

````python
def lowercase_vals(df: pd.DataFrame, type_val: str) -> pd.DataFrame:
    df_updated = df.copy()
    
    for col in df.select_dtypes(include=['object']).columns:
        df_updated[col] = df_updated[col].str.lower()
    
    return df_updated
````

Задание функции, переводящей числовые логические значения к строковым:

````python
def convert_vals(df: pd.DataFrame, col_names: List[str]) -> pd.DataFrame:
    df_updated = df.copy()
    
    for col in col_names:
        df_updated[col] = df_updated[col].apply(lambda x: 'yes' if x == 1 else 'no')
    
    return df_updated
````

#### 1.2.1 Числовые признаки

Задание функции, выводящей графики распределения значений в числовых столбцах:

````python
def plot_multi_box_hist(df: pd.DataFrame, 
                        cols_num: List[str], 
                        col_target: str,
                        title: str = 'Распределение значений числовых признаков в зависимости от целевой переменной\n') -> None:
    
    fig, ax = plt.subplots(2, len(cols_num), figsize=(12, 6))
    plt.suptitle(title, fontsize=16)
    
    for i, chart in enumerate(['box', 'hist']):
        for j, col in enumerate(cols_num):
            ax[i,j].set_title(col)

            target_vals = df[col_target].unique()
            target_dist = [df[df[col_target] == val][col] for val in target_vals]
            
            if chart == 'box':
                ax[i,j].boxplot(target_dist)
                ax[i,j].set_xticks(range(len(target_vals) + 2), ['', *target_vals, ''])
                legend_box = ax[i,j].legend(target_vals, framealpha=1).get_frame()

            if chart == 'hist':
                for n, dist in enumerate(target_dist):
                    ax[i,j].hist(dist, bins=30, alpha=0.8)
                    
                legend_box = ax[i,j].legend(target_vals, framealpha=1, loc='upper center').get_frame()
                
            legend_box.set_facecolor('white')
            legend_box.set_edgecolor('black')
                
            ax[i,j].grid()
    
    plt.tight_layout()
    plt.show()
````

#### 1.2.2 Категориальные признаки

Задание функции, строящей количественное распределение признаков:

````python
def plot_multi_bar(df: pd.DataFrame,
                   cols_cat: List[str],
                   n_cols: int = 4,
                   title: str = 'Распределение значений категориальных признаков\n\n\n') -> None:
    
    n_rows = len(cat_cols) // n_cols + len(cat_cols) % n_cols
    r, c = 0, 0

    fig, ax = plt.subplots(n_rows, n_cols, figsize=(12, n_rows * (n_cols + 1)))
    plt.suptitle(title, fontsize=16)
    
    plt_colors = list(colors._colors_full_map.values())

    for i, col in enumerate(cols_cat):        
        # get X labels and unique values count
        val_count = df[col].value_counts()
        labels = list(val_count.index)
        values = list(val_count.values)

        # plot bars
        ax[r,c].set_title(col)
        ax[r,c].bar(labels, values, color=plt_colors[:len(labels)], label=values)
        
        legend_box = ax[r,c].legend(framealpha=1).get_frame()
        legend_box.set_facecolor('white')
        legend_box.set_edgecolor('black')

        ax[r,c].grid()

        # rotate X axis ticks
        if any(len(str(n)) > 8 for n in labels):        
            for tick in ax[r,c].get_xticklabels():
                tick.set_rotation(90)

        # count rows and columns
        c = 0 if c == n_cols - 1 else c + 1
        r = r + 1 if c == 0 else r

    plt.tight_layout()
    plt.show()
````

## 2 Обучение моделей

Задание функции, строящей матрицу ошибок:

````python
def plot_corr_matrix(y_train: np.ndarray, 
                     y_pred: np.ndarray, 
                     target_col: str = target_col,  
                     title: str = '\nМатрица ошибок\nпредсказаний модели\n') -> None:
    
    plt.figure(figsize=(3, 2))
    plt.title(title, fontsize=11)
    
    variants = [f'not {target_col}', f'{target_col}']

    c_matrix = confusion_matrix(y_train, y_pred)
    c_matrix = pd.DataFrame(c_matrix, index=variants, columns=variants)
    
    sns.heatmap(
        c_matrix, 
        annot=True, 
        fmt='g',
        cmap='Blues',
        xticklabels=c_matrix.columns, 
        yticklabels=c_matrix.index
    )
    
    plt.xlabel('preds')
    plt.ylabel('real')
    plt.show()
````

Задание функции, сохраняющей результаты обучения моделей в набор данных:

````python
def save_results(save_to: pd.DataFrame, name: str, *args) -> pd.DataFrame:
    _score, _time, _model, _params = args
    
    save_to.at[name, 'score'] = _score
    save_to.at[name, 'time'] = _time
    save_to.at[name, 'model'] = _model
    
    save_to.at[name, 'params'] = None
    
    for col in ['params']:
        save_to[col] = save_to[col].astype('object')
    
    save_to.at[name, 'params'] = _params
    
    return save_to
````

Задание функции, преобразующей секунды в минуты:

````python
convert_time = lambda x: f'{np.round(x / 60, 2)} min' if x > 60 else f'{np.round(x, 2)} sec'
````

Задание функции, преобразующей значения словаря параметров:

````python
prettify_params = lambda x: {
    key.split('model__')[-1]: value for key, value in x.items()
}
````

Задание функции поиска лучшей модели:

````python
def search_model(estimator, 
                 X: pd.DataFrame,
                 y: pd.Series,
                 parameters: Dict, 
                 name: str,
                 save_to: pd.DataFrame,
                 metrics: str, 
                 n_cv: int) -> Tuple:
    
    # start timer    
    time_start = time.time()
    
    if len(parameters) > 3:
        search = RandomizedSearchCV(estimator, parameters, cv=n_cv, n_jobs=-1, 
                                    scoring=metrics, refit=True, verbose=0)
    else:
        search = GridSearchCV(estimator, parameters, cv=n_cv, n_jobs=-1, 
                              scoring=metrics, refit=True, verbose=0)

    search.fit(X, y)

    # save results
    best_score = search.best_score_
    best_model = search.best_estimator_
    best_params = prettify_params(search.best_params_)

    best_results = pd.DataFrame(search.cv_results_).loc[search.best_index_]
    best_time = convert_time(best_results['mean_fit_time'])
    
    # stop timer
    time_end = time.time() - time_start
    search_time = convert_time(time_end)
        
    return best_score, best_model, best_params, best_time, search_time
````

Задание функции обучения модели с подбором признаков:

````python
def train_model(estimator, 
                X: pd.DataFrame,
                y: pd.Series,
                parameters: Dict, 
                name: str,
                save_to: pd.DataFrame,
                metrics: str = 'roc_auc', 
                n_cv: int = 3) -> None:
    
    best_model = None
    
    print('Searching...')
    
    # get best model
    best_score, best_model, best_params, \
    best_time, search_time = search_model(estimator, X, y, parameters, 
                                          name, save_to, metrics, n_cv)
    
    print('Finished')
    print('Estimated search time:', search_time)
    
    # display params
    display(pd.DataFrame(best_params.values(), index=best_params.keys()).T)
    
    # save results
    save_to = save_results(save_to, name, best_score, best_time, best_model, best_params)
    
    # display results
    display(save_to.loc[name][['score', 'time']].to_frame().T)
    
    # plot correlation matrix
    plot_corr_matrix(y, best_model.predict(X))
````

### 2.7 Применение градиентного бустинга (2 балла)

**Задачи:**

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметроа количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию. 

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

---

Задание функции, объединяющей масштабированные числовые признаки с категориальными:

````python
def scale_num_and_concat(X_train_: pd.DataFrame,
                         X_valid_: pd.DataFrame,
                         X_test_: pd.DataFrame,
                         cols_num: List[str], 
                         cols_cat: List[str]) -> List[pd.DataFrame]:
    
    X_train_ = X_train_.reset_index(drop=True)
    X_valid_ = X_valid_.reset_index(drop=True)
    X_test_ = X_test_.reset_index(drop=True)
    
    X_train_scaled = num_pipe.fit_transform(X_train_[cols_num])
    X_valid_scaled = num_pipe.transform(X_valid_[cols_num])
    X_test_scaled = num_pipe.transform(X_test_[cols_num])
    
    results = []
    
    for X_init, X in [(X_train_, X_train_scaled), (X_valid_, X_valid_scaled), (X_test_, X_test_scaled)]:
        X_joined = pd.concat([pd.DataFrame(X, columns=cols_num), X_init[cols_cat]], axis=1)
        results.append(X_joined)
    
    return results
````

---

Масштабирование числовых признаков:

````python
X_train_scaled, X_valid_scaled, X_test_scaled = scale_num_and_concat(
    X_train, X_valid, X_test, num_cols, cat_cols
)
````

Задание подбираемых параметров:

````python
params_cbc = {
    'n_estimators': [900], # range(100, 2000, 100),
    'max_depth': [1], # range(1, 17),
    'learning_rate': [0.1], # [0.0001, 0.001, 0.01, 0.1],
    'l2_leaf_reg': [4], # np.arange(1, 11),
    'min_data_in_leaf': [700], # range(100, 1000, 100),
    'random_state': [RANDOM_STATE]
}
````

Обучение модели:

````python
train_model(
    estimator=CatBoostClassifier(cat_features=cat_cols, verbose=0), 
    X=X_train_scaled, 
    y=y_train, 
    parameters=params_cbc,
    name='cb_class', 
    save_to=results
)
````

>* Модель продемонстрировала истинно-положительный уклон - предсказано, что абонент не разорвал контракт, и это правда.
>* Модель ошибается в сторону ложно-отрицательных ответов - предсказано, что абонент разорвал контракт, и это ложь.

### 2.8 Применение стэкинга

Задание функции, объединяющей масштабированные числовые признаки с приведёнными к числовым категориальными:

````python
def scale_num_ohe_cat_features(X_train_: pd.DataFrame,
                               X_valid_: pd.DataFrame,
                               X_test_: pd.DataFrame,
                               cols_num: List[str], 
                               cols_cat: List[str]) -> List[pd.DataFrame]:
    
    X_train_scaled_ohe = full_pipe.fit_transform(X_train_)
    X_valid_scaled_ohe = full_pipe.transform(X_valid_)
    X_test_scaled_ohe = full_pipe.transform(X_test_)

    scaled_ohe_cols, results = [], []

    for transform, step, col in [
        ('num', 'scaler', cols_num), ('cat', 'ohe', cols_cat)
    ]:
        scaled_ohe_cols.extend(
            list(full_pipe.named_transformers_[transform] \
                          .named_steps[step] \
                          .get_feature_names_out(col))
        )
        
    for X in [X_train_scaled_ohe, X_valid_scaled_ohe, X_test_scaled_ohe]:
        results.append(pd.DataFrame(X, columns=scaled_ohe_cols))
    
    return results
````

---

Масштабирование числовых и приведение к числовым категориальных признаков:

````python
X_train_scaled_ohe, X_valid_scaled_ohe, X_test_scaled_ohe = scale_num_ohe_cat_features(
    X_train, X_valid, X_test, num_cols, cat_cols
)
````

Выведение на экран результатов обучения моделей:

````python
results[['score', 'time']].sort_values(by='score', ascending=False)
````

>Для стэкинга будут взяты модели с наилучшими значениями метрики ROC-AUC.

Задание пайплайна обучения модели:

````python
stacking = StackingClassifier(
    estimators=[
        ('log_reg', LogisticRegression(**results.loc['log_reg']['params'])),
        ('rf_class', RandomForestClassifier(**results.loc['rf_class']['params'])),
        ('cb_class', CatBoostClassifier(**results.loc['cb_class']['params'], verbose=0))
    ],
    final_estimator=CatBoostClassifier(verbose=0, random_state=RANDOM_STATE),
    n_jobs=-1,
)
````

Обучение модели:

````python
train_model(
    estimator=stacking, 
    X=X_train_scaled_ohe, 
    y=y_train, 
    parameters={}, 
    name='stacking', 
    save_to=results
)
````

>* Модель продемонстрировала истинно-положительный уклон - предсказано, что абонент не разорвал контракт, и это правда.
>* Модель ошибается в сторону ложно-отрицательных ответов - предсказано, что абонент разорвал контракт, и это ложь.

### 2.9 Выбор модели

Задание функции, строящей график ROC-кривой:

````python
def plot_roc_curve(estimator, X: pd.DataFrame, y: pd.Series, name: str) -> None:
    y_pred_proba = estimator.predict_proba(X)[:, 1]
    
    roc_auc = roc_auc_score(y, y_pred_proba)
    fpr_tpr_auc_roc = roc_curve(y, y_pred_proba)

    plt.plot(*fpr_tpr_auc_roc[:2], label='{} AUC={:.4f}'.format(name, roc_auc))
````

Задание функции, выводдящей графики ROC-кривых моделей с учётом обработки признаков:

````python
def plot_multi_roc_curves(save_to: pd.DataFrame,
                          X: Dict[str, pd.DataFrame],
                          y: pd.Series,
                          title: str = 'ROC-кривая моделей\n') -> None:
    
    plt.figure(figsize=(6, 5))
    plt.title(title)

    for name in results.index:
        model = save_to.loc[name]['model']

        if 'cb_class' in name:
            plot_roc_curve(model, X['cb_class'], y, name)
        elif 'stacking' in name:
            plot_roc_curve(model, X['stacking'], y, name)
        else:
            plot_roc_curve(model, X['x'], y, name)

    # random model
    plt.plot(np.linspace(0, 1, 100), np.linspace(0, 1, 100), linestyle='--', label='random')

    legend_box = plt.legend(framealpha=1).get_frame()
    legend_box.set_facecolor('white')
    legend_box.set_edgecolor('black')

    plt.grid()
    plt.show()
````

---

Выведение на экран результатов обучения моделей:

````python
results[['score', 'time']].sort_values(by='score', ascending=False)
````

Построение графика ROC-кривой:

````python
plot_multi_roc_curves(results, {'x': X_valid, 'cb_class': X_valid_scaled, 'stacking': X_valid_scaled_ohe}, y_valid)
````
